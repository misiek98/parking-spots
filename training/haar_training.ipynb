```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import os
```

```python
os.mkdir('pos')
os.mkdir('posInColor')
os.mkdir('neg')
os.mkdir('negInColor')
os.mkdir('xml_data')
```

```python
import shutil

for posImage in os.listdir('/content/posInColor'):
  img = cv2.imread(os.path.join('/content/posInColor', posImage))
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  cv2.imwrite(os.path.join('/content/pos', posImage), img)

shutil.rmtree('/content/posInColor')

for negImage in os.listdir('/content/negInColor'):
  img = cv2.imread(os.path.join('/content/negInColor', negImage))
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  cv2.imwrite(os.path.join('/content/neg', negImage), img)

shutil.rmtree('/content/negInColor')
```

```python
with open('neg.txt', 'a') as file:
  for negImage in os.listdir('/content/neg'):
    file.write(os.path.join('/content/neg', negImage))
    file.write('\n')
```
